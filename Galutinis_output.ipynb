```python
import pandas as pd
import numpy as np
import sys
import glob
from IPython.display import display
import os

# Nurodau directory is kur paimt failus
os.chdir('/home/jokubas/anaconda3/r/ms/contacts/asp_dna/')
# Atsifiltruoju tuscius failus, kuriuose nera kontaktu
txt_files = [f for f in os.listdir() if f.endswith('.txt') if os.stat(f).st_size > 77]
# Atrenku is failu baltymu id, pasidarau output dataframe, ir tuscia list
protein_ids = [ids.replace('_asp_dna_contacts.txt','') for ids in txt_files]
output=pd.DataFrame(columns=['ID','asp count','base count','contacts sum','sum for 1 asp','sum for 1 base','avg. contacts distance'])
dataframes_list = []
# Paleidziu for loop, kuriame apdoroju visus duomenu failus viena po kito ir istraukiu reikiama informacija bei sudelioju i output dataframe
for i in range(len(txt_files)):
    # Atidarau visus failus ir sudedu ju informacija i viena dataframe
    df = pd.read_csv('/home/jokubas/anaconda3/r/ms/contacts/asp_dna/'+txt_files[i])
    dataframes_list.append(df)
    #Atskiriu informacija ties > ir pasidarau stulpelius, taip pat juos pavadinu
    df = df.iloc[:,0].str.split('>',expand=True)    
    df.columns = ['Chain 1', 'Residue number', 'Atom number', 'Residue', 'Atom', 'Chain 2', 'Residue number 2', 'Atom Number2', 'Residue2', 'Atom2', 'Contacts and Distance']        
    df[['x','Contact area','Distance','s','j']]= df['Contacts and Distance'].str.split('\s', expand = True)
    # galutiniai stulpeliu duomenys
    pop = df[['Residue number', 'Residue', 'Atom', 'Residue number 2', 'Residue2', 'Atom2', 'Contact area','Distance']]
    # Stulpeliu duomenims priskiriu komandas, kurios suskaiciuoja mane dominancia info, viska paverciu i float, kad paskui galeciau suapvalinti
    aa_count = df['Residue number'].nunique()
    float(aa_count)
    base_count = df['Residue number 2'].nunique()
    float(base_count)
    df['Contact area'] = df['Contact area'].astype('float')
    contacts_sum = df['Contact area'].sum()
    float(contacts_sum)
    aa_contact = contacts_sum/aa_count
    float(aa_contact)
    base_contact = contacts_sum/base_count
    float(base_contact)
    protein_id = protein_ids[i]
    df['Distance'] = df['Distance'].astype('float')
    average_contacts_distance = df['Distance'].mean()
    results = {'ID':protein_id, 'asp count':aa_count,'base count': base_count,'contacts sum':round(contacts_sum, 2),'sum for 1 asp':round(aa_contact, 2),'sum for 1 base': round(base_contact, 2),'avg. contacts distance':round(average_contacts_distance, 2)}
    #rezultatus priskiriu output ir juos isspausdinu i txt faila, norodydamas jo directory
    output = output.append(results, ignore_index=True)
output.to_csv('/home/jokubas/anaconda3/output/MS/ms_output_asp_dna.txt', sep=('\t'), index=False)


    
    
    
    








    
  











```
